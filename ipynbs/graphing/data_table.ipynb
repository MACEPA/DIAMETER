```python
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('chained_assignment',None)
```

```python
all_files = glob.glob('C:/Users/lzoeckler/Desktop/qfu/binned/*.csv')
big_df = []
for file in all_files:
    df = pd.read_csv(file)
    df['num_data_points'] = len(df)
    big_df.append(df)
big_df = pd.concat(big_df, sort=True)
big_df.head()
```

```python
meta_df = big_df[['study_id', 'quansys_type', 'num_data_points']]
meta_df.drop_duplicates(inplace=True)
meta_df
```

```python
fourplex_df = meta_df.loc[meta_df['quansys_type'] == '4plex']
fourplex_df['all_points'] = fourplex_df['num_data_points'].sum()
fourplex_df['studies'] = len(fourplex_df)
fourplex_df = fourplex_df[['all_points', 'studies']]
fourplex_df.drop_duplicates(inplace=True)
fourplex_df
```

```python
fiveplex_df = meta_df.loc[meta_df['quansys_type'] == '5plex']
fiveplex_df['all_points'] = fiveplex_df['num_data_points'].sum()
fiveplex_df['studies'] = len(fiveplex_df)
fiveplex_df = fiveplex_df[['all_points', 'studies']]
fiveplex_df.drop_duplicates(inplace=True)
fiveplex_df
```
