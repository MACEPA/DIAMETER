```python
def clean_strings(val):
    clean = val.replace('<', '')
    clean = clean.replace('>', '')
    if clean == 'fail':
        return np.nan
    return clean
```

```python
standard = PdfPages('C:/Users/lzoeckler/Desktop/4plex/output_data/HRP2_groups_standard.pdf')
eighty = PdfPages('C:/Users/lzoeckler/Desktop/4plex/output_data/HRP2_groups_eighty.pdf')
less = PdfPages('C:/Users/lzoeckler/Desktop/4plex/output_data/HRP2_groups_less.pdf')
for pid in main_data['patient_id'].unique():
# for pid in ['pa-124']:
    pid_data = main_data.loc[main_data['patient_id'] == pid]
    plot_data = pid_data[['patient_id', 'time_point_days', 'HRP2_pg_ml',
                          'LDH_Pan_pg_ml', 'CRP_ng_ml', 'HRP2_pg_ml_dilution',
                          'HRP2_pg_ml_max_dilution']]
    plot_data['HRP2_pg_ml'] = plot_data['HRP2_pg_ml'].apply(clean_strings)
    plot_data['HRP2_pg_ml'] = plot_data['HRP2_pg_ml'].apply(float)
    plot_data['LDH_Pan_pg_ml'] = plot_data['LDH_Pan_pg_ml'].apply(clean_strings)
    plot_data['LDH_Pan_pg_ml'] = plot_data['LDH_Pan_pg_ml'].apply(float)
    plot_data['CRP_ng_ml'] = plot_data['CRP_ng_ml'].apply(clean_strings)
    plot_data['CRP_ng_ml'] = plot_data['CRP_ng_ml'].apply(float)
    first_days = plot_data['time_point_days'].unique().tolist()
    first_days.sort()
    first_days = first_days[:3]
    day_zero = first_days[0]
    initial_pg = plot_data.loc[plot_data['time_point_days'] == day_zero, 'HRP2_pg_ml'].item()
    plot_data['HRP2_pg_ml'] =  plot_data['HRP2_pg_ml'].apply(math.log)
    plot_data['LDH_Pan_pg_ml'] = plot_data['LDH_Pan_pg_ml'].apply(math.log)
    plot_data['CRP_ng_ml'] = plot_data['CRP_ng_ml'].apply(math.log)
    f = plt.figure()
    f.add_subplot()
    plt_color = 'blue'
    group = 1
    if initial_pg < 10:
        plt_color = 'black'
        group = 3
    if plt_color == 'blue':
        early_values = plot_data.loc[plot_data['time_point_days'].isin(first_days)]
        mean_val = early_values['HRP2_pg_ml'].mean()
        sub_plot = plot_data.loc[~plot_data['time_point_days'].isin(first_days)]
        sub_val = sub_plot['HRP2_pg_ml'].values
        for i in range(len(sub_val)):
            try:
                if (sub_val[i] - mean_val) > (.01 * mean_val):
                    if sub_val[i + 1] > .8 * mean_val:
                        plt_color = 'red'
                        group = 2
            except IndexError:
                pass
    if plt_color == 'red':
        print(pid)
    plt.plot(plot_data['time_point_days'], plot_data['HRP2_pg_ml'],
             color=plt_color)
    plt.plot(plot_data['time_point_days'], plot_data['LDH_Pan_pg_ml'],
            color='green')
    plt.plot(plot_data['time_point_days'], plot_data['CRP_ng_ml'],
            color='brown')
    plt.ylim(0, 20)
    plt.xlim(0, 130)
    pearson = pearsonr(plot_data['HRP2_pg_ml'], plot_data['LDH_Pan_pg_ml'])[0]
    spearman = spearmanr(plot_data['HRP2_pg_ml'], plot_data['LDH_Pan_pg_ml'])[0]
    pearson2 = pearsonr(plot_data['time_point_days'], plot_data['HRP2_pg_ml'])[0]
    spearman2 = spearmanr(plot_data['time_point_days'], plot_data['HRP2_pg_ml'])[0]
    pearson3 = pearsonr(plot_data['time_point_days'], plot_data['LDH_Pan_pg_ml'])[0]
    spearman3 = spearmanr(plot_data['time_point_days'], plot_data['LDH_Pan_pg_ml'])[0]
    title = "group: {}, patient_id: {},\nSpearman coefficient (time and HRP2): {},\nSpearman coefficient (time and Pan LDH): {}".format(group, pid, round(spearman2, 3), round(spearman3, 3))
    plt.title(title)
    plt.tight_layout()
#     if plt_color == 'red':
#         standard.savefig(f)
    standard.savefig(f)
    if max(plot_data['time_point_days']) >= 80:
        eighty.savefig(f)
    if (abs(spearman2) < .5) & (abs(spearman3) < .5):
        less.savefig(f)
    plt.close()
standard.close()
eighty.close()
less.close()
```
