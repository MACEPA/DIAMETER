```python
control_specific = pd.read_csv('{}/control_info.csv'.format(input_path))
main_cols = ['date', 'control_lot_id', 'kit_lot_id', 'analyzer_id',
             'temp', 'humidity', 'pass_fail', 'tech']
control_specific.head()
```

```python
hi_vals = control_specific[main_cols + ['hi_g6pd', 'hi_hb']]
hi_vals.rename(columns={'hi_g6pd': 'g6pd', 'hi_hb': 'hb', 'date': 'test_date'}, inplace=True)
hi_vals.head()
```

```python
low_vals = control_specific[main_cols + ['lo_g6pd', 'lo_hb']]
low_vals.rename(columns={'lo_g6pd': 'g6pd', 'lo_hb': 'hb', 'date': 'test_date'}, inplace=True)
low_vals.head()
```

```python
all_control_specific = pd.concat([hi_vals, low_vals])
all_control_specific.sort_values(['test_date', 'kit_lot_id', 'analyzer_id'], inplace=True)
all_control_specific = all_control_specific[['analyzer_id', 'test_date', 'g6pd', 'hb']]
all_control_specific.head()
```

```python
for_cs = red_cap.copy(deep=True)
for_cs = for_cs[['participant_id', 'cap_test_date', 'g6pd', 'hb']]
for_cs.rename(columns={'cap_test_date': 'test_date'}, inplace=True)
for_cs.head()
```

```python
cs_and_red = pd.concat([all_control_specific, for_cs])
cs_and_red.head()
```

```python
all_sdb = app.copy(deep=True)
all_sdb.head()
```

```python
sdb_and_cs_red = cs_and_red.merge(all_sdb, how='outer', on=['g6pd', 'hb'])
sdb_and_cs_red.head()
```

```python
sdb_and_cs_red.to_csv('{}/matched/total_red_compare.csv'.format(input_path), index=False)
```
