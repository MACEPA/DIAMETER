```python
import pandas as pd
```

```python
input_path = "C:/Users/lzoeckler/Desktop/G6PD_transcription/reduced"
prefix = "G6PD_reduced"
```

```python
click = pd.read_csv('{}/{}_clickmedix.csv'.format(input_path, prefix))
click = click[['patient_id', 'submit_date', 'start_date', 'end_date', 'time', 'id', 'temp', 'g6pd', 'hb']]
click.head()
```

```python
red = pd.read_csv('{}/{}_redcap.csv'.format(input_path, prefix))
red.head()
```

```python
red_cap = red[['participant_id', 'date', 'cap_test_date', 'cap_temp', 'cap_g6pd', 'cap_hb']]
red_cap.rename(columns={'date': 'cap_overall_date', 'cap_temp': 'temp', 'cap_g6pd': 'g6pd', 'cap_hb': 'hb'}, inplace=True)
red_cap.head()
```

```python
print(len(red_cap))
```

```python
app = pd.read_csv('{}/{}_0702_0712.csv'.format(input_path, prefix))
app = app[['g6pd', 'hb', 'date', 'control', 'device_n', 'sheet_id']]
app.head()
```

```python
print(len(app))
```

```python
no_control = app.loc[app['control'] == 0]
no_control.head()
```

```python
print(len(no_control))
```

```python
click_app = click.merge(no_control, how='outer', on=['g6pd', 'hb'])
click_app.head()
```

```python
click_app.to_csv('{}/click_app.csv'.format(input_path), index=False)
```

```python
red_app = red_cap.merge(no_control, how='outer', on=['g6pd', 'hb'])
red_app.head()
```

```python
red_app.tail()
```

```python
red_app.to_csv('{}/red_app.csv'.format(input_path), index=False)
```
