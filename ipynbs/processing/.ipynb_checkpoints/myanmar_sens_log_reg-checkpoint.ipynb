```python
import math
import glob
import numpy as np
import pandas as pd
from functools import reduce
from scipy.special import expit
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import confusion_matrix
from matplotlib.backends.backend_pdf import PdfPages
from sklearn.model_selection import train_test_split
```

```python
all_files = glob.glob('C:/Users/lzoeckler/Desktop/FOR_ML/binned/*.csv')
big_df = []
for file in all_files:
    df = pd.read_csv(file)
    big_df.append(df)
big_df = pd.concat(big_df, sort=True)
big_df = big_df.loc[big_df['study_id'] == '16-SMRU-01']
big_df.head()
```

```python
def clean_strings(val):
    if isinstance(val, str):
        if '<' in val:
            clean = 0
        elif '>' in val:
            clean = val.replace('>', '')
        try:
            return(np.log10(float(clean) + 0.1))
        except ValueError:
            return(np.nan)
    elif isinstance(val, float) or isinstance(val, int):
        return(np.log10(float(val) + 0.1))
    else:
        return(np.nan)
```

```python
val_cols = ['quansys_HRP2_pg_ml', 'quansys_LDH_Pan_pg_ml', 'quansys_LDH_Pv_pg_ml',
            'quansys_LDH_Pf_pg_ml', 'quansys_CRP_ng_ml']
fever = ['febrile']
```

```python
test_df = big_df.copy(deep=True)
test_df[val_cols] = test_df[val_cols].applymap(clean_strings)
q_vars = {'quansys_type': {'4plex': 0, '5plex': 1}, 'survey': {'longitudinal challenge': 0, 'health facility': 1,
                                                               'cross sectional': 2},
          'infection_category': {'uninfected': 0, 'nascent': 1, 'clearing': 2, 'symptomatic': 3, 'unifected': 0}}
test_df.replace(q_vars, inplace=True)
test_df['target'] = test_df['PCR_pos']
final_cols = val_cols + fever + ['target', 'quansys_type', 'study_id']
test_df = test_df[final_cols]
test_df = test_df[['quansys_HRP2_pg_ml', 'quansys_LDH_Pan_pg_ml', 'target']]
test_df.head()
```

```python
known_df = test_df.loc[~test_df['target'].isnull()]
hrp2_df = known_df.loc[~known_df['quansys_HRP2_pg_ml'].isnull()]
ldh_df = known_df.loc[~known_df['quansys_LDH_Pan_pg_ml'].isnull()]
```

```python
ldh_sens = []
ldh_spec = []
for i in range(1000):
    x_ldh_train, x_ldh_test, y_ldh_train, y_ldh_test = train_test_split(ldh_df['quansys_LDH_Pan_pg_ml'],
                                                    ldh_df['target'],
                                                    train_size=0.8,
                                                    test_size=0.2)
    
    ldh_data = x_ldh_train.values
    test_vals = y_ldh_train
    ldh_binary_data = test_vals.values.reshape(-1, 1)
    ldh_end_df = pd.DataFrame()
    
    ldh = linear_model.LogisticRegression(random_state=0)
    ldh.fit(ldh_data.reshape(-1, 1), ldh_binary_data)
    pred = x_ldh_test.values
    probabilities = ldh.predict_proba(pred.reshape(-1,1))
    non_prob = ldh.predict(pred.reshape(-1,1))
    ldh_end_df['probability_of_1_ldh'] = probabilities[:, 1]
    ldh_end_df['predicted_val_ldh'] = non_prob
    ldh_end_df['target'] = y_ldh_test.values
    
    target = ldh_end_df['target']
    results = ldh_end_df['predicted_val_ldh']
    cm = confusion_matrix(target, results)
    ldh_sens.append(cm[0,0] / (cm[0,0] + cm[0,1]))
    ldh_spec.append(cm[1,1] / (cm[1,0] + cm[1,1]))
```
