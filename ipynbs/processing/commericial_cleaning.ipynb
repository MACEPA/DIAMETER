```python
import pandas as pd
```

```python
com_data = pd.read_csv('C:/Users/lzoeckler/Desktop/table_pivot.csv')
com_data.rename(columns={'Quantity of 1ea/kit Standard G6PD Analyzer': 'Standard G6PD Analyzer',
                         'Quantity 25T/kit STANDARD G6PD Test / Ezi tube+': 'Standard G6PD Test/Ezi tube+',
                         '20T(10 low level, 10 high level/kit STANDARD G6PD Control Solution':
                             'Standard G6PD Control Solution'},
                inplace=True)
com_data.sort_values(['Purchase Date', 'Customer'], inplace=True)
com_data.head()
```

```python
check = pd.melt(com_data, id_vars=['Partner', 'Product Line', 'Purchase Date', 'Customer', 'Region', 'Country',
                                   'Remark', 'Distributor / Study'],
                var_name='Product Name', value_name='Units Sold')
check = check.loc[~check['Units Sold'].isnull()]
check.sort_values(['Purchase Date', 'Customer'], inplace=True)
```
