```python
import numpy as np
import pandas as pd
import seaborn as sns
import math
from scipy.stats import pearsonr, spearmanr
from sklearn.cluster import KMeans
from scipy.stats import pearsonr
from sklearn import linear_model
from sklearn.metrics import r2_score
from adjustText import adjust_text
from functools import partial
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
pd.set_option('chained_assignment', None)
from matplotlib.backends.backend_pdf import PdfPages
```

```python
main_data = pd.read_csv('C:/Users/lzoeckler/Desktop/4plex/output_data/final_dilutions.csv')
main_data.head()
```

```python
analyte_name_dict = {'HRP2_pg_ml': ('HRP2', 'pg/ml'), 'LDH_Pan_pg_ml': ('LDH_Pan', 'pg/ml'),
                     'LDH_Pv_pg_ml': ('LDH_Pv', 'pg/ml'), 'CRP_ng_ml': ('CRP', 'ng/ml')}
```

```python
analytes = ['HRP2_pg_ml', 'LDH_Pan_pg_ml', 'LDH_Pv_pg_ml', 'CRP_ng_ml']
```

```python
def clean_strings(val):
    clean = val.replace('<', '')
    clean = clean.replace('>', '')
    if clean == 'fail':
        return np.nan
    return clean
```

```python
all_pids = []
for pid in main_data['patient_id'].unique():
    oh_god = pd.DataFrame(data=[[pid]], columns=['patient_id'])
    pid_data = main_data.loc[main_data['patient_id'] == pid]
    plot_data = pid_data[['patient_id', 'time_point_days', 'HRP2_pg_ml',
                          'LDH_Pan_pg_ml', 'HRP2_pg_ml_dilution',
                          'HRP2_pg_ml_max_dilution']]
    plot_data['HRP2_pg_ml'] = plot_data['HRP2_pg_ml'].apply(clean_strings)
    plot_data['HRP2_pg_ml'] = plot_data['HRP2_pg_ml'].apply(float)
    first_days = plot_data['time_point_days'].unique().tolist()
    first_days.sort()
    first_days = first_days[:3]
    day_zero = first_days[0]
    initial_pg = plot_data.loc[plot_data['time_point_days'] == day_zero, 'HRP2_pg_ml'].item()
    if initial_pg >= 10:
        plot_data['HRP2_pg_ml'] = plot_data['HRP2_pg_ml'].apply(math.log)
        plot_data['LDH_Pan_pg_ml'] = plot_data['LDH_Pan_pg_ml'].apply(clean_strings)
        plot_data['LDH_Pan_pg_ml'] = plot_data['LDH_Pan_pg_ml'].apply(float)
        plot_data['LDH_Pan_pg_ml'] = plot_data['LDH_Pan_pg_ml'].apply(math.log)
        oh_god['h_l_p'] = pearsonr(plot_data['HRP2_pg_ml'], plot_data['LDH_Pan_pg_ml'])[0]
        oh_god['h_l_s'] = spearmanr(plot_data['HRP2_pg_ml'], plot_data['LDH_Pan_pg_ml'])[0]
        oh_god['t_h_p'] = pearsonr(plot_data['time_point_days'], plot_data['HRP2_pg_ml'])[0]
        oh_god['t_h_s'] = spearmanr(plot_data['time_point_days'], plot_data['HRP2_pg_ml'])[0]
        oh_god['t_l_p'] = pearsonr(plot_data['time_point_days'], plot_data['LDH_Pan_pg_ml'])[0]
        oh_god['t_l_s'] = spearmanr(plot_data['time_point_days'], plot_data['LDH_Pan_pg_ml'])[0]
    #     for timepoint in plot_data['time_point_days']:
    #         time_val = plot_data.loc[plot_data['time_point_days'] == timepoint,
    #                                  analyte].item()
    #         oh_god[timepoint] = time_val
        all_pids.append(oh_god)
all_pids = pd.concat(all_pids)
all_pids.dropna(inplace=True)
all_pids.head()
```

```python
kmeans = KMeans(n_clusters=2)
test = all_pids.drop('patient_id', axis=1)
kmeans.fit(test)
```

```python
output = kmeans.predict(test)
```

```python
output
```

```python
kmeans = all_pids.copy(deep=True)
kmeans['group'] = output
kmeans = kmeans[['patient_id', 'group']]
kmeans.head()
```

```python
pp = PdfPages('C:/Users/lzoeckler/Desktop/4plex/output_data/HRP2_groups_kmeans.pdf')
for pid in main_data['patient_id'].unique():
    pid_data = main_data.loc[main_data['patient_id'] == pid]
    plot_data = pid_data[['patient_id', 'time_point_days', 'HRP2_pg_ml',
                          'LDH_Pan_pg_ml', 'HRP2_pg_ml_dilution',
                          'HRP2_pg_ml_max_dilution']]
    plot_data['HRP2_pg_ml'] = plot_data['HRP2_pg_ml'].apply(clean_strings)
    plot_data['HRP2_pg_ml'] = plot_data['HRP2_pg_ml'].apply(float)
    plot_data['HRP2_pg_ml'] = plot_data['HRP2_pg_ml'].apply(math.log)
    plot_data['LDH_Pan_pg_ml'] = plot_data['LDH_Pan_pg_ml'].apply(clean_strings)
    plot_data['LDH_Pan_pg_ml'] = plot_data['LDH_Pan_pg_ml'].apply(float)
    plot_data['LDH_Pan_pg_ml'] = plot_data['LDH_Pan_pg_ml'].apply(math.log)
    group_df = kmeans.loc[kmeans['patient_id'] == pid]
    if len(group_df) > 0:
        group = group_df['group'].item()
    if group == 0:
        plt_color = 'blue'
    elif group == 1:
        plt_color = 'red'
    f = plt.figure()
    ax = f.add_subplot()
    plt.plot(plot_data['time_point_days'], plot_data['HRP2_pg_ml'], color=plt_color)
    plt.plot(plot_data['time_point_days'], plot_data['LDH_Pan_pg_ml'], color='green')
    title = "patient_id: {}\nkMeans group: {}".format(pid, group)
    plt.title(title)
    plt.tight_layout()
    pp.savefig(f)
    plt.close()
pp.close()
```
