```python
import numpy as np
import pandas as pd
import seaborn as sns
import math
from scipy.stats import pearsonr, spearmanr
from sklearn import linear_model
from sklearn.metrics import r2_score
from adjustText import adjust_text
from functools import partial
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
pd.set_option('chained_assignment',None)
from matplotlib.backends.backend_pdf import PdfPages
```

```python
# FOR GROUPS OF AT LEAST 4 DATA POINTS, STARTING AT THE INITIAL DATA POINT, DETERMINE SLOPE OF REGRESSION
# FIT TO SELECTED DATA POINTS. IF SLOPE IS EITHER POSITIVE OR CLOSE TO 0, CONTINUE SELECTING FOLLOWING
# DATA POINTS UNTIL SLOPE IS NO LONGER POSITIVE OR CLOSE TO 0, KEEP TRACK OF MAX NUMBER OF POINTS 
# (AKA BADDEST SECTION), 
```

```python
def clean_strings(val):
    clean = val.replace('<', '')
    clean = clean.replace('>', '')
    if clean == 'fail':
        return np.nan
    return clean
```

```python
main_data = pd.read_csv('C:/Users/lzoeckler/Desktop/4plex/output_data/final_dilutions.csv')
val_cols = ['HRP2_pg_ml', 'LDH_Pan_pg_ml', 'CRP_ng_ml']
main_data[val_cols] = main_data[val_cols].applymap(clean_strings)
main_data[val_cols] = main_data[val_cols].applymap(float)
main_data[val_cols] = main_data[val_cols].applymap(math.log)
main_data = main_data[['patient_id', 'time_point_days', 'HRP2_pg_ml',
                       'LDH_Pan_pg_ml', 'CRP_ng_ml', 'HRP2_pg_ml_dilution',
                       'HRP2_pg_ml_max_dilution']]
main_data.head()
```

```python
analyte_name_dict = {'HRP2_pg_ml': ('HRP2', 'pg/ml'), 'LDH_Pan_pg_ml': ('LDH_Pan', 'pg/ml'),
                     'LDH_Pv_pg_ml': ('LDH_Pv', 'pg/ml'), 'CRP_ng_ml': ('CRP', 'ng/ml')}
```

```python
def get_coef(df, time_vals):
    coef_data = df.loc[df['time_point_days'].isin(time_vals)]
    regr = linear_model.LinearRegression()
    time = coef_data['time_point_days'].values.reshape(-1,1)
    val = coef_data['HRP2_pg_ml'].values.reshape(-1,1)
    regr.fit(time, val)
    coef = np.float(regr.coef_)
    return coef
```

```python
good_df = []
bad_df = []
for pid in main_data['patient_id'].unique():
# for pid in ['pa-072']:
    pid_data = main_data.loc[main_data['patient_id'] == pid]
    pid_data.sort_values('time_point_days', inplace=True)
    all_times = pid_data['time_point_days'].unique().tolist()
    all_times.sort()
    max_run = 4
    the_rest = []
    i = 0
    end_val = 4
    baddest_section = []
    while (end_val <= len(all_times)) & (len(the_rest) != 1):
        time_vals = all_times[i:end_val]
        coef = get_coef(pid_data, time_vals)
        while (coef > -.01) & (len(time_vals) != 1) & (end_val < len(all_times)):
            end_val = end_val + 1
            time_vals = all_times[i:end_val]
            coef = get_coef(pid_data, time_vals)
            if coef > -.1:
                current_run = len(time_vals)
                if current_run > max_run:
                    max_run = current_run
                    baddest_section = time_vals
        i = end_val - 1
        try:
            all_times[i + 4]
            end_val = i + 4
        except:
            the_rest = all_times[i:]
            end_val = i + len(the_rest)
    good_vals = pid_data.loc[~pid_data['time_point_days'].isin(baddest_section)]
    good_df.append(good_vals)
    bad_vals = pid_data.loc[pid_data['time_point_days'].isin(baddest_section)]
    bad_df.append(bad_vals)
good_df = pd.concat(good_df)
bad_df = pd.concat(bad_df)
```

```python
time_series = PdfPages('C:/Users/lzoeckler/Desktop/4plex/output_data/funky_HRP2_timeseries.pdf')
for pid in main_data['patient_id'].unique():
# for pid in ['pa-124']:
    good_vals = good_df.loc[good_df['patient_id'] == pid]
    good_vals['group'] = 'blue'
    bad_vals = bad_df.loc[bad_df['patient_id'] == pid]
    bad_vals['group'] = 'red'
    combo = pd.concat([good_vals, bad_vals])
    combo = combo.sort_values('time_point_days')
    f = plt.figure()
    f.add_subplot()
    title = "patient_id: {}".format(pid)
    plt.plot(combo['time_point_days'], combo['HRP2_pg_ml'],
            c='black', alpha=0.6)
    plt.scatter(combo['time_point_days'], combo['HRP2_pg_ml'],
                c=combo['group'])
    plt.title(title)
    plt.xlabel('Time point, in days')
    plt.ylabel('Log of HRP2 pg/ml')
    plt.tight_layout()
    time_series.savefig(f)
    plt.close()
time_series.close()
```

```python
good_ratios = PdfPages('C:/Users/lzoeckler/Desktop/4plex/output_data/good_HRP2_vs_Pan.pdf')
f = plt.figure()
f.add_subplot()
regr = linear_model.LinearRegression()
time = good_df['HRP2_pg_ml'].values.reshape(-1,1)
val = good_df['LDH_Pan_pg_ml'].values.reshape(-1,1)
regr.fit(time, val)
pred = regr.predict(time)
score = r2_score(val, pred)
coef = np.float(regr.coef_)
intercept = np.float(regr.intercept_)
plt.scatter(good_df['HRP2_pg_ml'], good_df['LDH_Pan_pg_ml'], color='blue', alpha=0.6)
title = '"Good" data points, all patients\nk = {}'.format(round(coef, 4))
plt.plot(time, pred, color='black')
plt.title(title)
plt.xlim(0, 25)
plt.ylim(2.5, 20)
plt.xlabel('HRP2')
plt.ylabel('LDH Pan')
plt.tight_layout()
good_ratios.savefig(f)
# plt.close()
good_ratios.close()
```
